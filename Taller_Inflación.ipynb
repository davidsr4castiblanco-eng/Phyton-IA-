```python
paises = {
    "COL": "Colombia",
    "ECU": "Ecuador",
    "BRA": "Brasil",
    "CHL": "Chile",
    "MEX": "México",
    "PER": "Perú"
}

indicador = "FP.CPI.TOTL.ZG"
lista = []

for codigo, nombre in paises.items():
    url = f"https://api.worldbank.org/v2/country/{codigo}/indicator/{indicador}?format=json&date=2010:2023&per_page=100"
    respuesta = requests.get(url)
    data = respuesta.json()

    df_temp = pd.json_normalize(data[1])
    df_temp = df_temp[["date", "value"]].copy()
    df_temp["pais"] = nombre
    df_temp.columns = ["anio", "inflacion", "pais"]
    lista.append(df_temp)

inflacion = pd.concat(lista, ignore_index=True)
inflacion = inflacion[["pais", "anio", "inflacion"]]
inflacion["anio"] = pd.to_numeric(inflacion["anio"], errors="coerce")
inflacion = inflacion.sort_values(["pais", "anio"]).reset_index(drop=True)

inflacion.head()
```

```python
display(inflacion.head())
display(inflacion.tail(10))
display(inflacion.sample(min(3, len(inflacion)), random_state=42))
display(inflacion.iloc[1:5, [1]])

print("len:", len(inflacion))
print("shape:", inflacion.shape)
print("columns:", inflacion.columns)
print("dtypes:")
print(inflacion.dtypes)
inflacion.info()
```

```python
indicadores_alemania = {
    "NY.GDP.MKTP.KD.ZG": "crecimiento_pib",
    "NY.GDP.PCAP.PP.CD": "pib_percapita_ppa",
    "FP.CPI.TOTL.ZG": "inflacion",
    "SL.UEM.TOTL.ZS": "desempleo",
    "SP.POP.TOTL": "poblacion"
}

lista = []

for codigo_ind, nombre_ind in indicadores_alemania.items():
    url = f"https://api.worldbank.org/v2/country/DEU/indicator/{codigo_ind}?format=json&date=2010:2023&per_page=100"
    respuesta = requests.get(url)
    data = respuesta.json()

    df_temp = pd.json_normalize(data[1])
    df_temp = df_temp[["date", "value"]].copy()
    df_temp.columns = ["anio", nombre_ind]
    df_temp["anio"] = pd.to_numeric(df_temp["anio"], errors="coerce")
    lista.append(df_temp)

alemania = lista[0]
for df in lista[1:]:
    alemania = alemania.merge(df, on="anio", how="outer")

alemania = alemania.sort_values("anio").reset_index(drop=True)

alemania.head()
```

```python
display(alemania.head())
display(alemania.tail(10))
display(alemania.sample(min(3, len(alemania)), random_state=42))
display(alemania.iloc[1:5, [1]])

print("len:", len(alemania))
print("shape:", alemania.shape)
print("columns:", alemania.columns)
print("dtypes:")
print(alemania.dtypes)
alemania.info()
```
