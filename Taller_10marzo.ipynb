```python
import unicodedata

def limpiar_texto(texto):
    texto = texto.lower().strip()
    texto = unicodedata.normalize("NFKD", texto)
    texto = "".join(c for c in texto if not unicodedata.combining(c))
    return texto

def buscar_archivo(ruta_mes, texto):
    texto_limpio = limpiar_texto(texto)

    for archivo in ruta_mes.iterdir():
        nombre_limpio = limpiar_texto(archivo.name)
        if texto_limpio in nombre_limpio:
            return archivo

    disponibles = [a.name for a in ruta_mes.iterdir()]
    raise FileNotFoundError(f"No encontré un archivo que contenga: {texto}. Disponibles: {disponibles}")
```

```python
def leer_modulo(ruta_archivo):
    return pd.read_csv(
        ruta_archivo,
        low_memory=False,
        encoding="latin1",
        sep=";"
    )

def construir_base_mes(nombre_mes):
    ruta_mes = base_path / nombre_mes / "CSV"

    archivo_cg = buscar_archivo(ruta_mes, "generales")
    archivo_viv = buscar_archivo(ruta_mes, "hogar y la vivienda")
    archivo_ft = buscar_archivo(ruta_mes, "fuerza de trabajo")
    archivo_ocu = buscar_archivo(ruta_mes, "ocupados")

    print(f"\nProcesando {nombre_mes}")
    print("CG  ->", archivo_cg.name)
    print("VIV ->", archivo_viv.name)
    print("FT  ->", archivo_ft.name)
    print("OCU ->", archivo_ocu.name)

    cg = leer_modulo(archivo_cg)
    viv = leer_modulo(archivo_viv)
    ft = leer_modulo(archivo_ft)
    ocu = leer_modulo(archivo_ocu)

    base_mes = cg.merge(
        viv,
        on=["DIRECTORIO", "SECUENCIA_P"],
        how="left",
        suffixes=("", "_VIV")
    )

    base_mes = base_mes.merge(
        ft,
        on=["DIRECTORIO", "SECUENCIA_P", "ORDEN"],
        how="left",
        suffixes=("", "_FT")
    )

    base_mes = base_mes.merge(
        ocu,
        on=["DIRECTORIO", "SECUENCIA_P", "ORDEN"],
        how="left",
        suffixes=("", "_OCU")
    )

    base_mes["MES_NOMBRE"] = nombre_mes
    return base_mes
```

```python
meses = [
    "Julio 2025",
    "Agosto 2025",
    "Septiembre 2025",
    "Octubre 2025",
    "Noviembre 2025",
    "Diciembre 2025"
]

bases = []

for mes in meses:
    df_mes = construir_base_mes(mes)
    print("Shape:", df_mes.shape)
    bases.append(df_mes)
```
