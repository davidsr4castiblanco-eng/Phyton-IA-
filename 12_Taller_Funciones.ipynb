## **Ejercicio 4**

Un cajero automático tiene 500.000 pesos disponibles.
Un usuario puede retirar varias veces, pero no puede retirar más de lo que queda en el cajero. Escribe un programa que:

- Pregunte cuánto quiere retirar.

- Reste ese valor al saldo.

- Termine cuando el saldo llegue a 0 o el usuario escriba 0.

```python
saldo = 500000
print(f"Bienvenido al cajero automático. Saldo inicial: {saldo} pesos.")

while saldo > 0:
    try:
        retiro = float(input("¿Cuánto desea retirar? (Ingrese 0 para salir): "))

        if retiro == 0:
            print("Gracias por usar el cajero automático.")
            break

        if retiro < 0:
            print("El monto a retirar no puede ser negativo. Intente de nuevo.")
        elif retiro > saldo:
            print(f"Saldo insuficiente. Su saldo actual es: {saldo} pesos.")
        else:
            saldo -= retiro
            print(f"Ha retirado {retiro} pesos. Su nuevo saldo es: {saldo} pesos.")

    except ValueError:
        print("Entrada no válida. Por favor, ingrese un número.")

if saldo == 0:
    print("El cajero se ha quedado sin fondos.")
```

## **Ejercicio 5**

Crea una función raiz_cuadrada(num) que calcule la raíz cuadrada de un número.

Si el número es negativo, devuelve: "Error: no se puede calcular la raíz cuadrada de un número negativo."

Si el valor no es numérico, devuelve: "Error: el valor debe ser numérico."

NOTA: utiliza import math y la función math.sqrt(num). No olvides documentar la función.

```python
import math

def raiz_cuadrada(num):
    """
    Calcula la raíz cuadrada de un número.

    Args:
        num (int or float): El número del cual se calculará la raíz cuadrada.

    Returns:
        float or str: La raíz cuadrada del número, o un mensaje de error
                      si el número es negativo o no es numérico.
    """
    try:
        if not isinstance(num, (int, float)):
            return "Error: el valor debe ser numérico."
        if num < 0:
            return "Error: no se puede calcular la raíz cuadrada de un número negativo."
        return math.sqrt(num)
    except TypeError:
        return "Error: el valor debe ser numérico."
```

## **Ejercicio 6**

Escribe una función llamada es_par que reciba un número entero como parámetro y retorne True si el número es par y False si es impar.

NOTA: % es el operador módulo en Python. Lo que hace es dividir un número entre otro y devolver el residuo (lo que sobra de la división entera).

Si al dividir un número en 2, el residuo es cero, entonces el número es par, de lo contrario, es impar.

```python
def es_par(numero):
    """
    Verifica si un número entero es par.

    Args:
        numero (int): El número entero a verificar.

    Returns:
        bool: True si el número es par, False si es impar.
    """
    return numero % 2 == 0
```

## **Ejercicio 7**

Escribe un código que permita al usuario crear una contraseña y que verifique que esta cumpla todas las siguientes condiciones:

- Tiene al menos 8 caracteres.

- Contiene al menos una letra mayúscula.

- Contiene al menos una letra minúscula.

- Contiene al menos un número.

Si cumple todas las condiciones, muestra "Contraseña válida", de lo contrario, explica qué condición no se cumple.

```python
def validar_contrasena(contrasena):
    errores = []

    # Condición 1: Al menos 8 caracteres
    if len(contrasena) < 8:
        errores.append("La contraseña debe tener al menos 8 caracteres.")

    # Condición 2: Al menos una letra mayúscula
    if not any(char.isupper() for char in contrasena):
        errores.append("La contraseña debe contener al menos una letra mayúscula.")

    # Condición 3: Al menos una letra minúscula
    if not any(char.islower() for char in contrasena):
        errores.append("La contraseña debe contener al menos una letra minúscula.")

    # Condición 4: Al menos un número
    if not any(char.isdigit() for char in contrasena):
        errores.append("La contraseña debe contener al menos un número.")

    if not errores:
        return "Contraseña válida"
    else:
        return "\n".join(errores)


contrasena_usuario = input("Crea una contraseña: ")
resultado = validar_contrasena(contrasena_usuario)
print(resultado)
```

## **Ejercicio 10**

Vas conduciendo un carro muy rápido y eres detenido por un agente de tránsito.

Escribe una función "multa" que retorne uno de los 3 siguientes resultados:  

- Si la velocidad es 60 Km/h o menos, el resultado es "Sin multa".
- Si la velocidad está entre 60 y 80 Km/h (inclusive), el resultado es "Multa pequeña".
- Si la velocidad es mayor a 80 Km/h, el resultado es "Multa grande".

Además, considera que si esto ocurre el día de tu cumpleaños (codificado como parámetro booleano que recibe la función), tu velocidad podría ser 5 Km/ mayor en todos los casos.

```python
def multa(velocidad, es_cumpleanos):
    limite_sin_multa = 60
    limite_multa_pequena = 80

    if es_cumpleanos:
        limite_sin_multa += 5
        limite_multa_pequena += 5

    if velocidad <= limite_sin_multa:
        return "Sin multa"
    elif velocidad <= limite_multa_pequena:
        return "Multa pequeña"
    else:
        return "Multa grande"
```

```python
multa(81, True)
```

```python
multa(81, False)
```

## **Ejercicio 12**

Un país cobra impuestos según estos rangos:

- Menos de 2.000.000 → 0%

- Entre 2.000.000 y 5.000.000 → 10%

- Más de 5.000.000 → 20%

Escribe un programa que pida el ingreso anual de una persona y calcule cuánto debe pagar de impuestos.

```python
ingreso_anual = float(input("Ingrese su ingreso anual: "))

impuesto = 0

if ingreso_anual < 2000000:
    impuesto = 0
elif 2000000 <= ingreso_anual <= 5000000:
    impuesto = ingreso_anual * 0.10
else:
    impuesto = ingreso_anual * 0.20

print(f"Debe pagar {impuesto:,.2f} pesos de impuestos.")
```

## **Ejercicio 14**

Tienes un diccionario con exportaciones en millones de USD. Escribe un ciclo for que recorra los valores, calcule el promedio de exportaciones y lo imprima como: "El promedio de exportaciones es: XXXX millones de USD"

```python
exportaciones = {"Colombia": 1500, "México": 2300, "Chile": 1200, "Perú": 1800}
```

```python
total_exportaciones = 0
num_paises = 0

for exportacion in exportaciones.values():
    total_exportaciones += exportacion
    num_paises += 1

promedio_exportaciones = total_exportaciones / num_paises

print(f"El promedio de exportaciones es: {promedio_exportaciones:.2f} millones de USD")
```

## **Ejercicio 15**

Un estudiante invierte 1.000 dólares en un fondo que crece un 5% mensual. Escribe un programa que use while para calcular en cuántos meses su inversión supera los 2.000 dólares.

```python
inversion_inicial = 1000
tasa_crecimiento_mensual = 0.05
inversion_actual = inversion_inicial
meses = 0

while inversion_actual <= 2000:
    inversion_actual *= (1 + tasa_crecimiento_mensual)
    meses += 1

print(f"La inversión supera los 2000 dólares en {meses} meses.")
```

## **Ejercicio 16**

Una empresa quiere vender al menos 100 unidades de un producto. Cada día vende un número aleatorio de unidades entre 1 y 10 (pista: usa import random y random.randint(1,10)). Escribe un programa con while que muestre:

- El total vendido cada día.

- En cuántos días se logró la meta.

```python
import random

meta_ventas = 100
total_vendido = 0
dias = 0

print("Simulación de ventas diarias:")

while total_vendido < meta_ventas:
    dias += 1
    ventas_del_dia = random.randint(1, 10)  # Venta aleatoria entre 1 y 10 unidades
    total_vendido += ventas_del_dia
    print(f"Día {dias}: Se vendieron {ventas_del_dia} unidades. Total acumulado: {total_vendido} unidades.")

print(f"\nMeta alcanzada en {dias} días con un total de {total_vendido} unidades vendidas.")
```
